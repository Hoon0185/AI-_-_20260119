```python
import json

# Map Info
rows = 7
cols = 11
filename = "map_data_base.json"

# Create Map_List
my_map_list = [[False for _ in range(cols)] for _ in range(rows)]

# Tune Map Format
def format_row(row):
    # 각 요소를 JSON 형태의 문자열로 바꾼 뒤, 8칸 왼쪽 정렬 처리
    # f"{item:<8}"은 8칸을 확보하고 왼쪽( < )으로 정렬하라는 뜻입니다.
    formatted_items = [f"{json.dumps(item, ensure_ascii=False):<7}" for item in row]
    return "[" + ", ".join(formatted_items).rstrip() + "]"

# JSON 파일로 저장
try:
    with open(filename, 'w', encoding='utf-8') as f:
        f.write('[\n')  # 시작 괄호
        for i, row in enumerate(my_map_list):
            # 각 행을 JSON 문자열로 변환 (ensure_ascii=False는 한글/특수문자 방지)
            line = format_row(row)

            # 마지막 행이 아니면 뒤에 콤마(,) 추가
            if i < len(my_map_list) - 1:
                f.write(f'    {line},\n')
            else:
                f.write(f'    {line}\n')
        f.write(']')  # 닫는 괄호
        print(f"\n>>> [성공] {rows}x{cols} 크기의 맵이 '{filename}'으로 생성되었습니다.")

except Exception as e:
    print(f">>> [오류] 파일 저장 중 문제가 발생했습니다: {e}")
```
